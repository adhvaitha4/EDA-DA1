```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = {
    "Shopping Adviser": ["Advertising", "Friends/family", "Manufacturer websites", "News media", "online user reviews", "Retail websites", "Salespeople", "Other"],
    "Percentage": [7,45,5,11,13,4,1,14]
}
```

```python
df=pd.DataFrame(data1)
```

```python
plt.bar(df['Shopping Adviser'], df['Percentage'], color='skyblue')
plt.xlabel('Shopping Adviser')
plt.ylabel('Percentage')
plt.title('Bar Chart')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(df['Percentage'], labels=df['Shopping Adviser'],autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
df = df.sort_values(by='Percentage', ascending=False)
df['Cumulative Percentage'] = (df['Percentage'].cumsum() / df['Percentage'].sum()) * 10
```

```python
fig, ax1 = plt.subplots(figsize=(15, 8))
ax1.bar(df['Shopping Adviser'], df['Percentage'])
ax1.set_xlabel('Shopping Adviser')
ax1.set_ylabel('Percentage')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax2 = ax1.twinx()
ax2.plot(df['Shopping Adviser'], df['Cumulative Percentage'], marker='o', ms=7, linewidth=2)
ax2.set_ylabel('Cumulative Percentage (%)', )
ax2.tick_params(axis='y', labelcolor='red')
plt.title('Pareto Chart')
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Electricity cost data
data = [96, 171, 202, 178, 147, 102, 153, 197, 127, 82, 157, 185, 90, 116, 172, 141, 149, 206, 175,
        123, 128, 111, 148, 213, 130, 165, 144, 168, 109, 167, 130, 143, 187, 166, 139, 149, 129,
        158, 95, 163, 150, 154, 108, 119, 183, 151, 114, 135, 191, 137]

# Calculate histogram
counts, bins = np.histogram(data, bins=10)

# Calculate midpoints of bins
bin_midpoints = (bins[:-1] + bins[1:]) / 2

# Calculate percentage frequency
percentage_freq = 100 * counts / sum(counts)

# Plot percentage polygon
plt.figure(figsize=(10, 6))
plt.plot(bin_midpoints, percentage_freq, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.title("Percentage Polygon of Electricity Cost Distribution")
plt.xlabel("Electricity Cost")
plt.ylabel("Percentage Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data, [(i+1) / len(data) * 100 for i in range(len(data))], marker='o', label='Percentage')
plt.plot(data, [sum(data[:i+1]) / sum(data) * 100 for i in range(len(data))], marker='x', label='Cumulative Percentage')
plt.xlabel('Electricity Cost')
plt.ylabel('Percentage')
plt.title('Cumulative Percentage Polygon of Electricity Cost for One-Bedroom Apartments (July 2010)')
plt.legend()
plt.grid(True)
plt.show()
```
